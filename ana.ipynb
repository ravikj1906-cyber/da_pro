Clean the dataset 

```python
df.dropna(subset=['CustomerID', 'Description'], inplace=True)
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df.drop_duplicates(inplace=True)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.reset_index(drop=True, inplace=True)
print('Cleaned dataset shape:', df.shape)
df.head()
```

Top Selling Products (Maximum Selling Item)

```python
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

Products with Highest Revenue

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

top_revenue_products = df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
print(top_revenue_products)
```

Monthly Sales Analysis

```python
df['Month'] = df['InvoiceDate'].dt.month

monthly_sales = df.groupby('Month')['TotalPrice'].sum()
print(monthly_sales)
```

Country-wise Sales

```python
country_sales = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)
print(country_sales.head(10))
```

Top Customers (Highest Purchase)

```python
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

Most Frequently Bought Products

```python
most_frequent_products = df['Description'].value_counts().head(10)
print(most_frequent_products)
```

Average Order Value

```python
avg_order_value = df.groupby('InvoiceNo')['TotalPrice'].sum().mean()
print(avg_order_value)
```

Visualization of Monthly Sales

```python
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

Country-wise Sales Distribution

```python
country_sales.head(5).plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales Distribution by Country")
plt.ylabel("")
plt.show()
```
